```python
class triangle_dimensions:
    def __init__(self,x,y,z):
        self.a = x
        self.b = y
        self.c = z
        self.s = (x+y+z)/2
        
class triangle_area(triangle_dimensions):
    def __init__(self,*args):
        super(triangle_area,self).__init__(*args)
    def area(self):
        self.ar = (self.s*(self.s-self.a)*(self.s-self.b)*(self.s-self.c))**0.5
    
```

```python
a = triangle_area(8,10,5)
```

```python
a.area()
```

```python
a.ar
```

```python
def filter_long_words(obj,n):
    s = []
    for i in obj:
        if len(i)>n:
            s.append(i)
    return s
    
```

```python
l = ['abc','dfafd','fhdajfhd','hfasdjkhdfksdh','dsa','fhdkjha']
```

```python
filter_long_words(l,5)
```

```python
def map_words(obj):
    s = []
    for i in obj:
        s.append(len(i))
    return s
```

```python
map_words(l)
```

```python
def vowel_check(a):
    return a in ['a','e','i','o','u','A','E','I','O','U']
```

```python
vowel_check('a')
```

```python
vowel_check('A')
```

```python
vowel_check('z')
```
